```python
#-*- coding: utf-8 -*-
import numpy as np

class Perceptron(object): #定义感知器类
    '''
    eta : 学习率
    n_iter: 权重向量的训练次数
    w_: 神经分叉权重向量 W
    errors_ : 记录神经元判断出错次数
    '''
    def __int__(self,eta = 0.01,n_iter=10):  # initialize function 构造函数
        self.eta = eta
        self.n_iter = n_iter
        pass
    
    def fit(self,X,y):
        """
        输入训练数据，培训神经元，X输入样本向量，y对应样本分类
        
        X：shape(n_samples,n_features)
        X:[[1,2,3],[4,5,6]]
        n_samples: 2
        n_features: 3
        
        y:[1,-1] X 前分类为1，X后分类为-1
        """
        """
        初始化权重向量为0
        加一是因为前面算法提到的w0，也就是步调函数阈值
        """
        self.w = np.zero( 1+ X.shape[1])
        self.errors_ = []
        
        for _ in range(self.n_iter):
            errors_ = 0
            """
            X:[[1,2,3],[4,5,6]]
            y:[1,-1]
            zip(X,y) = [[1,2,3,1],[4,5,6,-1]]
            """
            for xi, target in zip(X,y):
                """
                update = η * (y-y')
                """
                update = self.eta * (target - self.predict(xi))
                
                """
                xi 是一个向量
                update * xi 等价：
                [Lw(1) = X[1]*update, Lw(2) = X[2]*update , ...] 
                """
                self.w_[1:] += update * xi
                self.w_[0] += update
                
                errors_ += int( update != 0.0)
                pass          
            pass
        
        def net_input(self,X):
            """
            z = w0*1 + w1*X1 +...+ wn*Xn"""
            return np.dot(X, self.w_[1:]) + self.w_[0]
            pass
        
        def predict(self, X):
            return np.where(self.net_input(X) >= 0.0 , 1,-1)
            pass
        
        pass
    
    
    
    
```
